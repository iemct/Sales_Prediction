```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
df=pd.read_csv('/home/dfx-rick/Downloads/sales_train.csv')
```

```python
df.shape
```

```python
df['date'].dtype
```

```python
df['years']= pd.to_datetime(df['date']).dt.strftime('%Y')
```

```python
df['months']= df.date.apply(lambda x: datetime.strptime(x, '%d.%m.%Y').strftime('%m'))
```

```python
df
```

```python
grouped= pd.DataFrame(df.groupby(['years', 'months'])['item_cnt_day'].sum().reset_index())
sns.pointplot(x='months', y='item_cnt_day', hue='years', data=grouped)
```

```python
g1=pd.DataFrame(df.groupby(['years' , 'months'])['item_price'].sum().reset_index())
sns.pointplot(x='months', y='item_price', hue='years', data=g1)
```

```python
g2=pd.DataFrame(df.groupby(['years' , 'shop_id'])['item_cnt_day'].mean().reset_index())
sns.lmplot(x='shop_id', y='item_cnt_day', hue='years', data=g2)
```
